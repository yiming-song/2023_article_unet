```python
import os

import h5py
import matplotlib.pyplot as plt
import numpy as np
import scipy
import spectrum_simulation

data_dir = os.environ.get("DATA_DIR") + "lidar_ml/real_data/simulation/"
```

```python
simulation_count = 10000

wind_profile_list = spectrum_simulation.get_random_wind_profile(simulation_count)
signal_list = spectrum_simulation.get_signal_spectrum(wind_profile_list)
noise_floor = spectrum_simulation.get_random_noise_floor(simulation_count)

total_signal = signal_list + noise_floor
```

```python
plt.plot(wind_profile_list[0])
plt.show()

plt.imshow(signal_list[0], vmin=-2, vmax=10)
plt.show()

plt.imshow(total_signal[0], vmin=-2, vmax=10)
plt.show()
```

```python
with h5py.File(data_dir + "100x100x10000_smooth=0.5_avg=0.hdf5", "w") as f:
    f["spectrum"] = total_signal
    f["spectrum_perfect"] = signal_list
    f["wind_profile"] = wind_profile_list
```
